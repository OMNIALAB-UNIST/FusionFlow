```python
import torch.utils.data
import sys
from os import path
import torchvision.transforms as transforms
sys.path.append(path.abspath("./pyfiles/"))
del sys.path[0]
# print(sys.path)
from RefurbishImitate import dataset
import customUtils
import RandAugment
import FastAutoAugment

traindir = '/data/1024batch/train'
batch_size=512

normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])
dataset = dataset.RefurbishDataset(
                        traindir, 
                        transforms.Compose([
                            RandAugment.augmentations.RandAugment(2, 9),
                            transforms.RandomResizedCrop(224),
                            transforms.ToTensor(),
                            normalize
                        ]),
                        cur_rank = 1,
                        cache_ratio = 0.5,
                        batch_size = batch_size
                    )

test_loader =  torch.utils.data.DataLoader(
                    dataset,
                    batch_size=batch_size, shuffle=False,
                    num_workers=1, pin_memory=True
                )

for image, target in test_loader:
    print(image.shape)
    print(sys.getsizeof(image[0]))
#     print(image, target)
```
